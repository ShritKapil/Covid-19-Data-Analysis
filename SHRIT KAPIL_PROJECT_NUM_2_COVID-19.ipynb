# converting datatypes

```python
covid['total_cases']=covid['total_cases'].astype(int)
covid['total_deaths']=covid['total_deaths'].astype(int)
covid['gdp_per_capita']=covid['gdp_per_capita'].astype(int)
```

```python
covid.drop(covid.columns[covid.columns.str.contains('unnamed',case=False)],axis=1,inplace=True)
```

# 	Top 10 Countries with Highest Covid Cases

```python
covid.groupby('location').agg({'total_cases':sum}).sort_values(by ='total_cases',ascending=False).head(10)
```

```python
c=covid.groupby('location').agg({'total_cases':sum}).sort_values(by ='total_cases',ascending=False).head(10).plot(kind='bar',figsize=(10,5),title='Top 10 Countries with Highest Covid Cases')
```

# Total number of deaths among the countries with the highest number of covid-19 cases

```python
covid.groupby('location').agg({'total_cases':sum,'total_deaths':sum}).sort_values(by ='total_cases',ascending=False).head(10)
```

```python
covid.groupby('location').agg({'total_cases':sum,'total_deaths':sum}).sort_values(by ='total_cases',ascending=True).head(10).plot(kind='line',figsize=(10,5),title='Total number of deaths among the countries')
```

# total covid-19 cases monthwise

```python
#introducing or extracting new column (month)and (year) from date
covid['date']=pd.to_datetime(covid['date'],format='%Y%m%d')
covid['year']=pd.DatetimeIndex(covid['date']).year
covid['month']=pd.DatetimeIndex(covid['date']).month
                                                                                       
```

```python
covid
```

```python
covid.groupby(['year','month'])['total_cases'].sum()
```

```python
covid.groupby(['year','month'])['total_cases'].sum().plot(kind='bar',figsize=(10,6),title='month-wise total cases')
plt.legend()
plt.xlabel('month')
plt.ylabel('total_cases')
plt.show()
```

# Total covid-19 cases datewise

```python
covid.groupby('date')['total_cases'].sum()
```

```python
covid.groupby('date')['total_cases'].sum().head(40).plot(kind='bar',figsize=(10,5),title='Total covid-19 cases datewise')
plt.legend()
plt.ylabel('total_cases')
plt.grid()
```
